```python
#get the Trend

import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

nifty_50_stocks = [
    ('TCS', 'IT'),
    ('INFY', 'IT'),
    ('HCLTECH', 'IT'),
    ('TECHM', 'IT'),
    ('WIPRO', 'IT'),
    ('TATASTEEL', 'Metals'),
    ('HINDALCO', 'Metals'),
    ('JSWSTEEL', 'Metals'),
    ('COALINDIA', 'Mining'),
    ('ONGC', 'Energy - Oil & Gas'),
    ('IOC', 'Energy - Oil & Gas'),
    ('RELIANCE', 'Conglomerate'),
    ('BHARTIARTL', 'Telecommunications'),
    ('IDEA', 'Telecommunications'),
    ('POWERGRID', 'Utilities - Power'),
    ('NTPC', 'Utilities - Power'),
    ('ADANIPORTS', 'Infrastructure'),
    ('LT', 'Infrastructure'),
    ('SBIN', 'Financial Services - Banking'),
    ('HDFCBANK', 'Financial Services - Banking'),
    ('ICICIBANK', 'Financial Services - Banking'),
    ('AXISBANK', 'Financial Services - Banking'),
    ('KOTAKBANK', 'Financial Services - Banking'),
    ('BAJFINANCE', 'Financial Services - NBFC'),
    ('M&M', 'Automobile'),
    ('MARUTI', 'Automobile'),
    ('HEROMOTOCO', 'Automobile'),
    ('BAJAJ-AUTO', 'Automobile'),
    ('TATAMOTORS', 'Automobile'),
    ('DRREDDY', 'Pharmaceuticals'),
    ('CIPLA', 'Pharmaceuticals'),
    ('SUNPHARMA', 'Pharmaceuticals'),
    ('DIVISLAB', 'Pharmaceuticals'),
    ('HINDUNILVR', 'FMCG - Consumer Goods'),
    ('ITC', 'FMCG - Consumer Goods'),
    ('NESTLEIND', 'FMCG - Consumer Goods'),
    ('BRITANNIA', 'FMCG - Consumer Goods'),
    ('ASIANPAINT', 'FMCG - Consumer Goods'),
    ('ULTRACEMCO', 'Cement'),
    ('SHREECEM', 'Cement'),
    ('GRASIM', 'Cement'),
    ('TITAN', 'Consumer Durables'),
    ('INDUSINDBK', 'Financial Services - Banking'),
    ('JSWENERGY', 'Utilities - Power'),
    ('UPL', 'Chemicals'),
    ('TATACONSUM', 'FMCG - Consumer Goods'),
    ('SBILIFE', 'Financial Services - Insurance'),
    ('HDFCLIFE', 'Financial Services - Insurance'),
    ('ADANIENT', 'Conglomerate')
]

nifty_next_50_stocks=[
    ('ABB', 'ABB India Ltd.', 'Capital Goods'),
    ('ADANIENSOL', 'Adani Energy Solutions Ltd.', 'Power'),
    ('ADANIGREEN', 'Adani Green Energy Ltd.', 'Power'),
    ('ATGL', 'Adani Total Gas Ltd.', 'Oil Gas & Consumable Fuels'),
    ('AWL', 'Adani Wilmar Ltd.', 'Fast Moving Consumer Goods'),
    ('AMBUJACEM', 'Ambuja Cements Ltd.', 'Construction Materials'),
    ('DMART', 'Avenue Supermarts Ltd.', 'Consumer Services'),
    ('BAJAJHLDNG', 'Bajaj Holdings & Investment Ltd.', 'Financial Services'),
    ('BANKBARODA', 'Bank of Baroda', 'Financial Services'),
    ('BERGEPAINT', 'Berger Paints India Ltd.', 'Consumer Durables'),
    ('BEL', 'Bharat Electronics Ltd.', 'Capital Goods'),
    ('BOSCHLTD', 'Bosch Ltd.', 'Automobile and Auto Components'),
    ('CANBK', 'Canara Bank', 'Financial Services'),
    ('CHOLAFIN', 'Cholamandalam Investment and Finance Company Ltd.', 'Financial Services'),
    ('COLPAL', 'Colgate Palmolive (India) Ltd.', 'Fast Moving Consumer Goods'),
    ('DLF', 'DLF Ltd.', 'Realty'),
    ('DABUR', 'Dabur India Ltd.', 'Fast Moving Consumer Goods'),
    ('GAIL', 'GAIL (India) Ltd.', 'Oil Gas & Consumable Fuels'),
    ('GODREJCP', 'Godrej Consumer Products Ltd.', 'Fast Moving Consumer Goods'),
    ('HAVELLS', 'Havells India Ltd.', 'Consumer Durables'),
    ('HAL', 'Hindustan Aeronautics Ltd.', 'Capital Goods'),
    ('ICICIGI', 'ICICI Lombard General Insurance Company Ltd.', 'Financial Services'),
    ('ICICIPRULI', 'ICICI Prudential Life Insurance Company Ltd.', 'Financial Services'),
    ('IOC', 'Indian Oil Corporation Ltd.', 'Oil Gas & Consumable Fuels'),
    ('IRCTC', 'Indian Railway Catering And Tourism Corporation Ltd.', 'Consumer Services'),
    ('NAUKRI', 'Info Edge (India) Ltd.', 'Consumer Services'),
    ('INDIGO', 'InterGlobe Aviation Ltd.', 'Services'),
    ('JINDALSTEL', 'Jindal Steel & Power Ltd.', 'Metals & Mining'),
    ('LICI', 'Life Insurance Corporation of India', 'Financial Services'),
    ('MARICO', 'Marico Ltd.', 'Fast Moving Consumer Goods'),
    ('MUTHOOTFIN', 'Muthoot Finance Ltd.', 'Financial Services'),
    ('PIIND', 'PI Industries Ltd.', 'Chemicals'),
    ('PIDILITIND', 'Pidilite Industries Ltd.', 'Chemicals'),
    ('PGHH', 'Procter & Gamble Hygiene & Health Care Ltd.', 'Fast Moving Consumer Goods'),
    ('PNB', 'Punjab National Bank', 'Financial Services'),
    ('SBICARD', 'SBI Cards and Payment Services Ltd.', 'Financial Services'),
    ('SRF', 'SRF Ltd.', 'Chemicals'),
    ('MOTHERSON', 'Samvardhana Motherson International Ltd.', 'Automobile and Auto Components'),
    ('SHREECEM', 'Shree Cement Ltd.', 'Construction Materials'),
    ('SHRIRAMFIN', 'Shriram Finance Ltd.', 'Financial Services'),
    ('SIEMENS', 'Siemens Ltd.', 'Capital Goods'),
    ('TVSMOTOR', 'TVS Motor Company Ltd.', 'Automobile and Auto Components'),
    ('TATAPOWER', 'Tata Power Co. Ltd.', 'Power'),
    ('TORNTPHARM', 'Torrent Pharmaceuticals Ltd.', 'Healthcare'),
    ('TRENT', 'Trent Ltd.', 'Consumer Services'),
    ('MCDOWELL-N', 'United Spirits Ltd.', 'Fast Moving Consumer Goods'),
    ('VBL', 'Varun Beverages Ltd.', 'Fast Moving Consumer Goods'),
    ('VEDL', 'Vedanta Ltd.', 'Metals & Mining'),
    ('ZOMATO', 'Zomato Ltd.', 'Consumer Services'),
    ('ZYDUSLIFE', 'Zydus Lifesciences Ltd.', 'Healthcare')
]


def percentageIncrease(stock_data):
  # Calculating the average stock value for the first 5 days
  average_first_5_days = stock_data.head(5)['Close'].mean()
  # Getting the maximum stock value from the rest of the dates
  max_value_rest_of_days = stock_data['Close'][5:].max()
  # Calculating the percentage hike w.r.t the first 5 days avg value
  percentage_increase = ((max_value_rest_of_days - average_first_5_days) / average_first_5_days) * 100
  # print(f"The avg value of the stock for the first 5 days: {average_first_5_days:.2f}")
  # print(f"The maximum value of the stock from the rest of the dates: {max_value_rest_of_days:.2f}")
  # print(f"The percentage hike w.r.t the first 5 days avg value: {percentage_increase:.2f}%")
  return (percentage_increase)

def analyze_Trend(symbol):
    print('------Analyze--------:',symbol)
    # Fetching stock data for the past 7 days for the ticker 'AAPL'
    stock_data = yf.download(symbol+'.NS', period='30d')['Close']

    # Resetting index to have the date as a column
    stock_data = stock_data.reset_index()
    stock_data_last_7=stock_data[-7:]
    stock_data_last_14=stock_data[-14:]

    # Plotting the stock values with regression line
    plt.figure(figsize=(10, 6))
    sns.set(style='whitegrid')

    ax = sns.regplot(x=stock_data.index, y='Close', data=stock_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
    ax = sns.regplot(x=stock_data_last_7.index, y='Close', data=stock_data_last_7, line_kws={'color': 'green'})
    ax = sns.regplot(x=stock_data_last_14.index, y='Close', data=stock_data_last_14, line_kws={'color': 'black'})

    ax.set_xlabel('Days')
    ax.set_ylabel('Stock Value')
    ax.set_title('Stock Values with Regression Line for Past 7 Days')

    # Calculating the slope of the regression line
    slope, intercept, r_value, p_value, std_err = stats.linregress(stock_data.index, stock_data['Close'])
    print("Last 30 days Slope value of the regression line:", slope)

    # Calculating the slope of the regression line
    slope_14, intercept_14, r_value_14, p_value_14, std_err_14 = stats.linregress(stock_data_last_14.index, stock_data_last_14['Close'])
    print("Last 14 days Slope value of the regression line:", slope_14)

    # Calculating the slope of the regression line
    slope_7, intercept_7, r_value_7, p_value_7, std_err_7 = stats.linregress(stock_data_last_7.index, stock_data_last_7['Close'])

    print("Last 7 days Slope value of the regression line:", slope_7)
    #
    percentageIncrease_value=percentageIncrease(stock_data)

    plt.tight_layout()
    plt.show()

    return ((symbol,percentageIncrease_value,slope,slope_14,slope_7))


dict_symbols={}
for symbol in nifty_next_50_stocks:
  dict_symbols[symbol]=analyze_Trend(symbol[0])
sorted_data = dict(sorted(dict_symbols.items(), key=lambda item: item[1][1]))
print("\n======================================================================\n")

# Calculate the maximum length of the keys
max_key_length = max(len(str(key)) for key in sorted_data.keys())

# Printing each key-value pair with formatted output
for key, value in sorted_data.items():
    print(f"{str(key).ljust(max_key_length)}: {value}")



```
